```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
print(cv2.__version__)
```

```python
# HISTOGRAMA P&B
# O eixo X do gráfico normalmente possui uma distribuição de 0 a 255 que demonstra o valor (intensidade) do pixel
# E no eixo Y é plotada a qtde de pixels daquela intensidade

img = cv2.imread('flores.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img, cmap="gray")

#Função calcHist para calcular o histograma da imagem
h = cv2.calcHist ([img], [0], None, [256], [0, 256])
plt.figure()
plt.title("Histograma P&B")
plt.xlabel("Intensidade")
plt.ylabel("Qtde de Pixels")
plt.plot(h)
plt.hist(img.ravel(),256,[0,256]) #usando a função ravel para plotar
plt.show()
cv2.waitKey(0)
```
