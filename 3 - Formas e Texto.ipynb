```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
print(cv2.__version__)
```

```python
imagem = np.zeros(shape=[50,50, 3], dtype=np.uint8)
plt.imshow(imagem)
plt.title("Canvas")
## criando quadrado 50x50 p adicionar funções de cor rgb.
```

```python
imagem[0, 0] = (0, 255, 0) #Pintando o ponto 0x0 de verde.
imagem[30:50, :] = (255,0,0)# Pintando entre os pontos 30x50 de vermelho.
plt.imshow(imagem)
```

```python
imagem2 = np.zeros(shape=[500,500, 3], dtype=np.uint8)
plt.imshow(imagem2)
plt.title("500x500")
## criando quadrado 500x500 p criar obj.
```

```python
imagem2[30:50, :] = (255, 0, 0) #Cria um retangulo vermelho por toda a largura da imagem
imagem2[100:150, 50:100] = (0, 0, 255) #Cria um  quadrado azul
imagem2[:, 200:220] = (0, 255, 255)#Cria um retangulo ciano
imagem2[150:300, 250:350] = (0, 255, 0)#Cria um retangulo verde da linha 150 a 300 nas colunas 250 a 350
imagem2[300:400, 50:150] = (255, 255, 0)#Cria um quadrado amarelo da linha 150 a 300 nas colunas 250 a 350
imagem2[250:350, 300:400] = (255, 255, 255) #Cria um quadrado branco
plt.imshow(imagem2)
plt.title("500x500")
```

```python
#Com a técnica de slicing é possível criar quadrados e retângulos.
#Para outras formas geométricas é possível utilizar as funções da OpenCV, isso é útil principalmente no caso de desenho de círculos e textos sobre a imagem.

vermelho = (0,0, 255)
verde = (0,255,0)
azul = (255,0,0)

cv2.line(imagem2, (0,0,), (100,200), verde)
cv2.line(imagem2, (300,200,), (150,150), vermelho, 5)
cv2.rectangle(imagem, (20, 20), (120, 120), azul, 10)
cv2.rectangle(imagem, (200, 50), (225, 125), verde, -1)

(x, y) = (imagem2.shape[1] // 2, imagem2.shape[0] // 2)
for raio in range(0, 175, 15):
    cv2.circle(imagem2, (y,x), raio, vermelho)
    plt.imshow(imagem2)
```

```python
imagem3 = np.zeros(shape=[500,500, 3], dtype=np.uint8)
plt.imshow(imagem3)
## criando quadrado 500x500 p escrever texto.


## Outra função útil é a de escrever textos sobre a imagem.
fonte = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(imagem3,'OpenCV',(15,65),fonte,2,(255,255,255),2,cv2.LINE_AA) #Escrevendo OpenCV nas coordenadas descritas
plt.imshow(imagem3)
plt.title("500x500")
```
