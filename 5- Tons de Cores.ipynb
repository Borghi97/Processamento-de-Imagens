```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
print(cv2.__version__)
```

```python
#Alterando tons de cor da imagem original
img = cv2.imread("narutinho.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.title("Imagem Original")

```

```python
img = cv2.imread("narutinho.jpg")
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(hsv)
plt.title("Imagem em tom de hsv")
```

```python
img = cv2.imread("narutinho.jpg")
lab = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
plt.imshow(lab)
plt.title("Imagem em tom de L*a*b")
```

```python
img5 = cv2.imread("narutinho.jpg")
gray = cv2.cvtColor(img5, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap = 'gray')
plt.title("Imagem em tom de cinza")
```
