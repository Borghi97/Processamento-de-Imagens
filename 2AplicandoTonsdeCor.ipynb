```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
print(cv2.__version__)
```

```python
imagem = np.zeros(shape=[50,50, 3], dtype=np.uint8)
plt.imshow(imagem)

## criando quadrado 50x50 p adicionar funções de cor rgb.
```

```python
for y in range(0, imagem.shape[0]):#percorrendo linhas
    for x in range(0, imagem.shape[1]):#percorrendo colunas
        imagem[y,x] = (255,0,0)
        plt.imshow(imagem)
        #percorrendo a imagem e adicionando pixel vermelho
```

```python
for y in range(0, imagem.shape[0]): #percorrendo linhas
    for x in range(0, imagem.shape[1]):#percorrendo colunas
        imagem[y,x] = (x%256,y%256,x%256) #utilizando a operação resta da divisao por 256 para manter o resultado entre 0 e 255.
        plt.imshow(imagem) #mostrando resultado
```

```python
for y in range(0, imagem.shape[0]): #percorrendo linhas
    for x in range(0, imagem.shape[1]): #percorrendo colunas
        imagem[y,x] = (0, (x*y)%256, 0) #Utilizando valores de linha multiplicado pela coluna (x*y) no componente "g"
        plt.imshow(imagem)
```
