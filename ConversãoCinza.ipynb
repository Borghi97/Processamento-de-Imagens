```python
import cv2
from matplotlib import pyplot as plt

print(cv2.__version__)

#Usando matplotlib p conversão de imagem p tons de cinza
```

```python
img = cv2.imread("ceu.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Importando a img e convertendo o formato padrão p escala de cinza
```

```python
plt.imshow(gray, cmap = 'gray')
plt.title("Céu cinza")
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Céu colorido")

#Convertendo formato padrão para rgb e printando imagem original
```
