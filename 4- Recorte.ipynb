```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import imutils
print(cv2.__version__)
```

```python
#Usando técnica para recortar a imagem

imagem = cv2.imread("flores.jpg")
imagem = cv2.cvtColor(imagem, cv2.COLOR_BGR2RGB)
plt.imshow(imagem)
recorte = imagem[100:500, 100:500]

f = plt.figure(figsize=(15, 15))
f.add_subplot(1,2,1)
plt.imshow(recorte)
plt.title("Recorte")
```
